```python
import pandas as pd
```

```python
df = pd.read_excel("sample_transitions.xlsx", dtype={'Plant':'str', 'Channel': 'str', 'from_material': 'str', 'to_material': 'str'})
```

```python
df.head()
```

```python
df.columns.tolist()
```

```python
# Collect all unique materials (from + to) for each group
group_cols = ['Plant', 'Channel', 'DP_Group_L3', 'Material_Hierarchy', 'Material_Desc']

# A cleaner approach that avoids the deprecated .append:
transition_materials = (
    pd.concat([df[group_cols + ['from_material']].rename(columns={'from_material': 'material'}),
               df[group_cols + ['to_material']].rename(columns={'to_material': 'material'})])
      .groupby(group_cols)['material']
      .apply(lambda x: sorted(x.unique().tolist()))
      .rename('transition_materials_group')
      .reset_index()
)

transitions_df = df.merge(transition_materials, on=group_cols, how='left')
```
